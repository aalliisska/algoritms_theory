```python
def pairStability(hospitalsPrefer, studentsPrefer, pairs):
    studentHospital = {}
    firstPair = list(pairs.keys())[0]
    print(firstPair)
    
    if firstPair in hospitalsPrefer:
        # {больница: студент}
        for hospital, student in pairs.items():
            studentHospital[student] = hospital
    else:
        # {студент: больница}
        studentHospital = pairs.copy()
    
    # {больница: студент} 
    hospitalStudent = {}
    for student, hospital in studentHospital.items():
        hospitalStudent[hospital] = student
    
    for hospital in hospitalsPrefer:
        for student in studentsPrefer:
            if studentHospital.get(student) == hospital:
                continue
            
            pairStudentHospital = hospitalStudent.get(hospital)
            currentHospitalStudent = studentHospital.get(student)
            
            hospitalList = hospitalsPrefer[hospital]
            newStudentPosition = hospitalList.index(student)
            currentStudentPosition = hospitalList.index(pairStudentHospital)

            studentList = studentsPrefer[student]
            thisHospitalPosition = studentList.index(hospital)
            studentHospitalPosition = studentList.index(currentHospitalStudent)
            
            if newStudentPosition < currentStudentPosition and thisHospitalPosition < studentHospitalPosition:
                return "Неустойчивое"
    
    return "Устойчивое"


hospitalsPrefer = {
    "Больница 1": ["Иванов", "Петров", "Сидоров"],
    "Больница 2": ["Петров", "Сидоров", "Иванов"],
    "Больница 3": ["Сидоров", "Иванов", "Петров"]
}

studentsPrefer = {
    "Иванов": ["Больница 1", "Больница 2", "Больница 3"],
    "Петров": ["Больница 2", "Больница 1", "Больница 3"],
    "Сидоров": ["Больница 3", "Больница 1", "Больница 2"]
}


pairs = {
    "Больница 1": "Иванов",
    "Больница 2": "Петров",
    "Больница 3": "Сидоров"
}
print("Тест 1: ", pairStability(hospitalsPrefer, studentsPrefer, pairs))

raspredelenie2 = {
    "Больница 1": "Петров",
    "Больница 2": "Иванов",
    "Больница 3": "Сидоров"
}
print("Тест 2: ", pairStability(hospitalsPrefer, studentsPrefer, raspredelenie2))

raspredelenie3 = {
    "Иванов": "Больница 2",
    "Петров": "Больница 1",
    "Сидоров": "Больница 3"
}
print("Тест 3: ", pairStability(hospitalsPrefer, studentsPrefer, raspredelenie3))
```

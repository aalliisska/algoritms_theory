```python
from collections import deque

def check_consistency(n, same_pairs, diff_pairs):
    total_nodes = n + len(same_pairs)
    graph = [[] for _ in range(total_nodes)] 
    
    helper_node = n
    
    for sample1, sample2 in same_pairs:
        v = helper_node
        helper_node += 1
        
        graph[sample1].append((v, 1))
        graph[v].append((sample1, 1))
        graph[sample2].append((v, 1))
        graph[v].append((sample2, 1))
    
    for sample1, sample2 in diff_pairs:
        graph[sample1].append((sample2, 1))
        graph[sample2].append((sample1, 1))
    
    colors = [-1] * helper_node  
    
    for start_node in range(helper_node):
        if colors[start_node] != -1:
            continue 
        
        queue = deque([start_node])
        colors[start_node] = 0
        
        while queue:
            current = queue.popleft()
            
            for neighbor, color_diff in graph[current]:
                expected_color = colors[current] ^ color_diff
                
                if colors[neighbor] == -1:
                    colors[neighbor] = expected_color
                    queue.append(neighbor)
                elif colors[neighbor] != expected_color:
                    print(f"Конфликт на узле {current}")
                    return False
    return True



print(check_consistency(3, [(0,1)], [(1,2)]))
print(check_consistency(3, [(0,1), (1,2)], [(0,2)]))
    
```

```python
from collections import deque

def is_bipartite(edges, n):
    graph = [[] for _ in range(n)]
    for u, v in edges:
        graph[u].append(v)
        graph[v].append(u)

    colors = [-1] * n
    
    for start in range(n):
        if colors[start] == -1:
            queue = deque([start])
            colors[start] = 0
            
            while queue:
                u = queue.popleft()
                
                for v in graph[u]:
                    if colors[v] == -1:
                        colors[v] = 1 - colors[u]
                        queue.append(v)
                    elif colors[v] == colors[u]:
                        return False
    
    return True


def find_perfect_matchings(edges, n):
    if not is_bipartite(edges, n):
        return []
    
    if n % 2 != 0:
        return []
    
    graph = [[] for _ in range(n)]
    for u, v in edges:
        graph[u].append(v)
        graph[v].append(u)
    
    left_size = n // 2
    perfect_matchings = []
    
    def backtrack(matching, used_right, index):
        if index == left_size:
            perfect_matchings.append(matching.copy())
            return
        
        for neighbor in graph[index]:
            if not used_right[neighbor]:
                used_right[neighbor] = True
                matching.append((index, neighbor))
                backtrack(matching, used_right, index + 1)
                matching.pop()
                used_right[neighbor] = False
    
    used_right = [False] * n
    backtrack([], used_right, 0)
    
    return perfect_matchings


def main():
    edges = [(0, 3), (0, 4), (1, 3), (1, 5), (2, 4), (2, 5)]
    n = 6
    
    print(f"Ребра: {edges}")
    print(f"Двудольный: {'да' if is_bipartite(edges, n) else 'нет'}")
    
    matchings = find_perfect_matchings(edges, n)
    if matchings:
        for i, matching in enumerate(matchings, 1):
            print(f"  {i}. {matching}")
    else:
        print("  нет")


if __name__ == "__main__":
    main()
```
